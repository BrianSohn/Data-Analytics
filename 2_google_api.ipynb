```python
import requests
import json
from time import sleep
from random import random
```

```python
api_key = 'your_api_key'
```

```python
# example query 
query = 'restaurants near Columbia University'
```

```python
def get_businesses(query, api_key, min_rating=0.0, max_results=30):
    query = query.replace(' ', '%20')
    url="https://maps.googleapis.com/maps/api/place/textsearch/json?query=%s&key=%s" % (query,api_key)
    result_list = []
    result_count = 0
    
    try:
        response = requests.get(url)
        if not response.status_code == 200:
            print("HTTP error:", response.status_code)
        else: 
            try: 
                response_data = response.json()
                results = response_data.get('results')
                
                for result in results:
                    if (result.get('rating') != None) and (result.get('rating') > min_rating): 
                        name = result.get('name')
                        formatted_address = result.get('formatted_address')
                        opening_hours = result.get('opening_hours')
                        price_level = result.get('price_level')
                        rating = result.get('rating')

                        result_list.append((name, formatted_address, opening_hours, price_level, rating))
                        result_count += 1
                    if result_count >= max_results:
                        break
                
                
                while (result_count < max_results) and (response_data.get('next_page_token')):
                    new_url = "https://maps.googleapis.com/maps/api/place/textsearch/json?pagetoken=%s&key=%s" % (response_data.get('next_page_token'), api_key)
                    response = requests.get(new_url)
                    while response.json().get('status') != 'OK':
                        sleep(random())
                        response = requests.get(new_url)
                    try: 
                        response_data = response.json()
                    except:
                        print("Response not in valid JSON format")
                        response_data = {}
                        results = None
                        
                    results = response_data.get('results')
                
                    for result in results:
                        if (result.get('rating') != None) and (result.get('rating') > min_rating): 
                            name = result.get('name')
                            formatted_address = result.get('formatted_address')
                            opening_hours = result.get('opening_hours')
                            price_level = result.get('price_level')
                            rating = result.get('rating')

                            result_list.append((name, formatted_address, opening_hours, price_level, rating))
                            result_count += 1
                        if result_count >= max_results:
                            break
  
                return result_list

            except:
                print("Response not in valid JSON format")
    except: 
        print("Something went wrong with requests.get")
```

```python
get_businesses(query, api_key, min_rating = 4.0, max_results = 30)
```
